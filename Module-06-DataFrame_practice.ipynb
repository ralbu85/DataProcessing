# 데이터프레임 실전

## 1. 원데이터 (csv 데이터)

실전에서는 누구도 여러분에게 바로 데이터프레임을 제공하지 않는다. 여러분은 거의 대부분의 상황에서 텍스트파일을 공들여 수집하고, 수많은 전처리를 거치고 나서야 pandas 모듈로 데이터프레임을 불러들여 여러가지 분석을 할 수 있을 것이다. 이런 원 데이터를 효과적으로 수집하는 방법은 차후 다루기로 하고, 여기서는 데이터를 처리하기 전 텍스트 파일을 얻었다고 가정한다.

여러분이 처리하게 될 대부분의 데이터는 csv 파일로 되어 있다. csv(comma seperated value), 말그대로 콤마형태로 구분된 텍스트 형태의 파일이다. 이번 챕터에서는 csv파일을 pandas로 불러들여와서 보다 실질적인 분석을 해 보도록 하겠다. 우선 csv 파일의 생김새를 살펴보자. 공유폴더에 업로드 되어있는 olympics.csv 파일에는 올림픽에 출전한 국가와 메달의 수가 써있다. 더블클릭하여 살펴보자.

맨 위 줄에는 데이터의 헤드가 써있고, 그다음줄부터 국가별 메달 획득 실적이 써있다. 

## 2. padnas 데이터 불러오기

이 데이터를 pandas의 데이터프레임으로 불러들이기 위해서는 read_csv 명령어를 사용하면 된다.

```python
import pandas as pd
df=pd.read_csv('olympics.csv')
```

```python
df
```

보다시피 위에서 콤마로 나뉘어져 있던 부분을 자동으로 깔끔하게 정리하여 데이터프레임으로 변환되어 있는 것을 볼 수 있다. 아직 완전한 전처리가 끝나지 않았지만, 이를 바탕으로 여러가지 데이터 분석을 할 수 있는 희망이 보인다.

## 5. 데이터 살펴보기

### 5.1 하계올림픽에서 금메달을 한개 이상 딴 국가들만 추려보고 전체 국가중 몇퍼센트를 차지하는가?

```python
only_gold=df[df['Gold']>0]
only_gold.head()
```

## 7. 인구데이터 분석

```python
df = pd.read_csv('census.csv')
df.head()
```

### unique 메소드 (해당 칼럼에 유일한 값이 몇개 들어있는가?)

```python
df['CTYNAME'].unique()
```

```python
df['SUMLEV'].unique() # SUMLEV은 지역코드를 의미
```

```python
df=df[df['SUMLEV']==50] # 지역코드가 50인 데이터만 추려내기
df.head()
```

### 칼럼만 추려내기

```python
columns_to_keep = ['STNAME',
                   'CTYNAME',
                   'BIRTHS2010',
                   'BIRTHS2011',
                   'BIRTHS2012',
                   'BIRTHS2013',
                   'BIRTHS2014',
                   'BIRTHS2015',
                   'POPESTIMATE2010',
                   'POPESTIMATE2011',
                   'POPESTIMATE2012',
                   'POPESTIMATE2013',
                   'POPESTIMATE2014',
                   'POPESTIMATE2015']
df = df[columns_to_keep]
df.head()
```

### 이차원 인덱스

```python
df = df.set_index(['STNAME','CTYNAME'])
df.head()
```

### 이차원인덱스에서 데이터 고르기

```python
df.loc['Michigan','Washtenaw County']
```
