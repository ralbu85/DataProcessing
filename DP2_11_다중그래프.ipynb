## 다중 그래프 만들기 방법 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10,0.1)
np.cos(x)
```

```python
plt.figure(figsize=(8,4))
# 첫번째 그래프
plt.subplot(1,3,1) # 1행 3열, 그 중 첫번째
plt.plot(x,np.sin(x),color='red',label='sin(x)')
plt.title('sin curve')
plt.legend()
# 두번째 그래프
plt.subplot(1,3,2)
plt.plot(x,np.cos(x),color='green') #1행 3열, 그중 두번째
plt.title('cos curve')
# 세번째 그래프
plt.subplot(1,3,3) #1행 3열중 세번째 그래프
plt.scatter(x,np.sin(x)+2)
plt.title('scatter sin curve')
```

## 다중그래프 만들기 2번째 방법 (정석)

```python
fig, axes= plt.subplots(1,3,figsize=(8,4)) #fig 전체 그림, #axes 개별 그래프
axes[0].plot(x,np.sin(x),c='red',label='sin(x)')
axes[0].legend()
axes[1].plot(x,np.cos(x),c='green',label='cos(x)')
axes[1].legend()
axes[2].scatter(x,np.sin(x)+2)
```

## 다중그래프 만들기 (축공유)

```python
fig, axes= plt.subplots(1,3,figsize=(8,4),sharey=True) #fig 전체 그림, #axes 개별 그래프
axes[0].plot(x,np.sin(x),c='red',label='sin(x)')
axes[0].legend()
axes[1].plot(x,np.cos(x),c='green',label='cos(x)')
axes[1].legend()
axes[2].scatter(x+3,np.sin(x+3)+2)
```

## 다수의 그래프를 알고리즘적으로 그려보기

```python
fig,axes=plt.subplots(3,3,figsize=(8,8),sharex=True,sharey=True)
for i in range(3):
    for j in range(3):
        if (i+j)%2==0:
            axes[i,j].plot(x,np.sin(x),color='red')
        else:
            axes[i,j].plot(x,np.cos(x),color='green')
    
```

## 다수의 히스토그램 그려보기

```python
fig, axes = plt.subplots(2,2,sharex=True,figsize=(8,6))
for i in range(0,len(fig.axes)):
    sample_size=10**(i+1) # 10,100,1000,10000
    sample = np.random.normal(size=sample_size)
    axes[i//2, i%2].hist(sample,bins=100)
    axes[i//2, i%2].set_title(str(sample_size))

```
