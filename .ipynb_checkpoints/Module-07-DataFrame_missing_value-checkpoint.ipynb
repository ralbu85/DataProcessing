# 결측치 처리

```python
import pandas as pd
import numpy as np
```

```python
data= pd.Series([1,np.nan,'Hello', None])
data
```

```python
data.isnull()
```

```python
data[data.isnull()]
```

```python
data[data.notnull()]
```

## 결측치 제거전략1: 지우기

```python
data.dropna()
```

```python
df = pd.DataFrame([[1,      np.nan, 2],
                   [2,      3,      5],
                   [np.nan, 4,      6]])
df
```

```python
df.dropna(axis='columns')
```

```python
df.dropna(axis='rows')
```

```python
df[3]=np.nan
df
```

```python
df.dropna(axis='columns',how='all')
```

```python
df.dropna(axis='columns',how='any')
```

```python
df.dropna(axis='rows',thresh=3)
```

## 결측치 제거전략 2: 다른값으로 채우기

```python
data = pd.Series([1, np.nan, 2, None, 3], index=list('abcde'))
data
```

```python
data.fillna(0)
```

```python
data.fillna(method='ffill')
```

```python
data.fillna(method='bfill')
```
